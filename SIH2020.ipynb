```python
import math
import numpy
fileName = "temp.csv"
file = open(fileName)
data = file.read()
dataList = []
dataLineWise = list(data.split("\n"))
del(dataLineWise[-1])
for i in dataLineWise:
    dataList.append(tuple(map(float,i.split(", "))))
velocityList = []
for i in range(len(dataList)-1):
    velocityX = (dataList[i+1][0]-dataList[i][0])/(dataList[i+1][3]-dataList[i][3])
    velocityY = (dataList[i+1][1]-dataList[i][1])/(dataList[i+1][3]-dataList[i][3])
    velocityZ = (dataList[i+1][2]-dataList[i][2])/(dataList[i+1][3]-dataList[i][3])
    velocity = math.sqrt(pow(velocityX,2)+pow(velocityY,2)+pow(velocityZ,2))
    velocityList.append((velocityX,velocityY,velocityZ,velocity,dataList[i][3]))
file.close()
velocityFileName = "velocity.txt"
velocityFile = open(velocityFileName,"w")
for velocityData in velocityList:
    velocityFile.write(str(velocityData[0])+", "+str(velocityData[1])+", "+str(velocityData[2])+", "+str(velocityData[3])+", "+str(velocityData[4])+"\n")
velocityFile.close()
```

```python
import math
#for finding velocity angle change wrt X-Y Plane
inputFileName = "velocity.txt"
outputFileName = "angularChange.txt"
data = readFile(inputFileName)
outFile = open(outputFileName,"w")
for i in range(len(data)-1):
    x1 = data[i][0]
    y1 = data[i][1]
    x2 = data[i+1][0]
    y2 = data[i+1][1]
    dotProduct = x1*x2 + y1*y2
    denom = (math.sqrt(math.pow(x1,2)+math.pow(y1,2)))*(math.sqrt(math.pow(x2,2)+math.pow(y2,2)))
    angle = math.acos(dotProduct/denom)
    print(angle)
    outFile.write(str(angle)+", "+str(data[i][4])+"\n")
outFile.close()
```

```python
def readFile(fileName):
    file = open(fileName,"r")
    data = file.read()
    dataList = []
    dataLineWise = list(data.split("\n"))
    del(dataLineWise[-1])
    for i in dataLineWise:
        dataList.append(tuple(map(float,i.split(", "))))
    file.close()
    return dataList
counter = 0
DIR = "C:\\Users\\Dell\\Desktop\\Data\\"
lis = ["CivillianAirplane","CivillianHelicopter","FighterAirplane","FighterHelicopter","IRBM"]
lis1 = ["Trajectory","Velocity","Acceleration","AngularChangeRate"]
dic = {'v':0,'vz':0,'a':0,'delTheta':0,'z':0}
for i in range(len(lis)):
    outFileName = "statistics"
    fout = open(DIR+lis[i]+outFileName,"w")
    for fileName in lis1:
        inFileName = DIR+lis[i]+"\\"+fileName
        data = readFile(inFileName)
        if fileName == "Trajectory":
            for tup in data:
                dic['z']+=tup[2]
                counter += 1
            dic['z']/=counter
            counter = 0
        if fileName == "Velocity":
            for tup in data:
                dic['v']+=tup[3]
                dic['vz']+=tup[2]
                counter += 1
            dic['v']/=counter
            dic['vz']/=counter
            counter = 0
        if fileName == "Acceleration":
            for tup in data:
                dic['a']+=tup[3]
                counter += 1
            dic['a']/=counter
            counter = 0
        if fileName == "AngularChangeRate":
            for tup in data:
                dic['delTheta']+=tup[0]
                counter += 1
            dic['delTheta']/=counter
            counter = 0          
    
```
