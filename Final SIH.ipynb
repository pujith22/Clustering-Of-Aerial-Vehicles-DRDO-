```python
import random
import numpy
import math
no_of_files = 1000

# Vmax, Vmin for velocity of ranges.
V_max = 450
V_min = 200

# point of time between the range Vmax and Vmin in our model
# Vz for maximum altitudinal velocity that can be attained
# Vz_min and Vz_max for altitudinal velocity ranges(in KM/Hr).
Vz_min = -200
Vz_max = 200

# Ranges of altitude of the object at which the object can fly (in KM).
Altitude_Max = 10
Altitude_Min = 0.15

# pitchAngleRate that can be achieved in radians/sec.
PitchAngleRate = 0.3

# maxacceleration that can be achieved in KM/HR*2
max_acceleration = 80000
#acceleration_rate = 1500
velocityLimiter = 1
vzLimiter = 1
acceleration_initial = 35000
vz_initial = 0
pAngleRate = 0
pAngle = 0.7854
```

```python
def secToHour(sec):
    return sec/3600
DT = secToHour(0.05)
DIR = "C:\\Users\\Dell\\Desktop\\Data\\FighterHelicopter\\Trajectory\\"
for fileNum in range(1,no_of_files+1):
    # creating new Files
    outFileName = DIR+str(fileNum)+".csv"
    outFile = open(outFileName,"w")
    # initializing initial Parameters.
    x = 0
    y = 0
    z = random.random()*(Altitude_Max-Altitude_Min)+Altitude_Min
    v_initial = random.randint(V_min,V_max)
    V = v_initial
    a = acceleration_initial
    Vz = vz_initial    
    #simulating the trajectory
    for time in numpy.arange(0,20,0.05):
        newAcceleration = a + random.random()*10 - 10   
        while(newAcceleration>max_acceleration):
            newAcceleration -= acceleration_rate
        newVelocity = V + a*DT
        while(not(newVelocity>=V_min and newVelocity<=V_max)):
            if (newVelocity>V_max):
                newVelocity -= velocityLimiter
            elif (newVelocity<V_min):
                newVelocity += velocityLimiter
        newVz = Vz + random.random()*vzLimiter - (vzLimiter/2)
        while(not(newVz>=Vz_min and newVz<=Vz_max)):
            if (newVz>Vz_max):
                newVz -= vzLimiter
            elif (newVz<Vz_min):
                newVz += vzLimiter
        #print(str(V)+", "+str(Vz))
        Vxy = math.sqrt(V**2 - Vz**2)
        new_pAngleRate = pAngleRate + random.random()*2*PitchAngleRate - PitchAngleRate
        new_pAngle = pAngle + pAngleRate*0.05
        Vx = Vxy/math.sqrt(1+math.tan(pAngle)**2)
        Vy = Vx * math.tan(pAngle)
        newX = x + Vx*DT
        newY = y + Vy*DT
        newZ = z + Vz*DT
        outFile.write(str(x)+", "+str(y)+", "+str(z)+", "+str(time)+"\n")
        x = newX
        y = newY
        z = newZ
        a = newAcceleration
        V = newVelocity
        Vz = newVz
        pAngle = new_pAngle
        pAngleRate = new_pAngleRate
    outFile.close()
```

```python
import math
import numpy
no_of_files = 1000
extension = ".csv"
directory = "C:\\Users\\Dell\\Desktop\\Data\\FighterHelicopter\\"
for fileNum in range(1,no_of_files+1):
    fileName = directory+"Trajectory\\"+str(fileNum)+extension
    file = open(fileName,'r')
    data = file.read()
    dataList = []
    dataLineWise = list(data.split("\n"))
    del(dataLineWise[-1])
    for i in dataLineWise:
        dataList.append(tuple(map(float,i.split(", "))))
    velocityList = []
    for i in range(len(dataList)-1):
        velocityX = (dataList[i+1][0]-dataList[i][0])/(dataList[i+1][3]-dataList[i][3])
        velocityY = (dataList[i+1][1]-dataList[i][1])/(dataList[i+1][3]-dataList[i][3])
        velocityZ = (dataList[i+1][2]-dataList[i][2])/(dataList[i+1][3]-dataList[i][3])
        velocity = math.sqrt(pow(velocityX,2)+pow(velocityY,2)+pow(velocityZ,2))
        velocityList.append((velocityX,velocityY,velocityZ,velocity,dataList[i][3]))
    file.close()
    velocityFileName = directory + "Velocity\\"+str(fileNum)+ extension
    velocityFile = open(velocityFileName,"w")
    for velocityData in velocityList:
        velocityFile.write(str(velocityData[0])+", "+str(velocityData[1])+", "+str(velocityData[2])+", "+str(velocityData[3])+", "+str(velocityData[4])+"\n")
    velocityFile.close()
```

```python
import math
import numpy
no_of_files = 1000
extension = ".csv"
directory = "C:\\Users\\Dell\\Desktop\\Data\\FighterHelicopter\\"
for fileNum in range(1,no_of_files+1):
    fileName = directory+"Velocity\\"+str(fileNum)+extension
    file = open(fileName,'r')
    data = file.read()
    dataList = []
    dataLineWise = list(data.split("\n"))
    del(dataLineWise[-1])
    for i in dataLineWise:
        dataList.append(tuple(map(float,i.split(", "))))
    List = []
    for i in range(len(dataList)-1):
        x1 = dataList[i][0]
        y1 = dataList[i][1]
        x2 = dataList[i+1][0]
        y2 = dataList[i+1][1]
        dotProduct = x1*x2 + y1*y2
        denom = (math.sqrt(math.pow(x1,2)+math.pow(y1,2)))*(math.sqrt(math.pow(x2,2)+math.pow(y2,2)))
        temp = dotProduct/denom
        if temp>1:
            temp = 1
        angle = math.acos(temp)
        List.append((angle,dataList[i][4]))
    file.close()
    FileName = directory + "PitchAngle\\"+str(fileNum)+ extension
    File = open(FileName,"w")
    for Data in List:
        File.write(str(Data[0])+", "+str(Data[1])+"\n")
    File.close()
```

```python
import math
import numpy
no_of_files = 1000
extension = ".csv"
directory = "C:\\Users\\Dell\\Desktop\\Data\\FighterHelicopter\\"
for fileNum in range(1,no_of_files+1):
    fileName = directory+"PitchAngle\\"+str(fileNum)+extension
    file = open(fileName,'r')
    data = file.read()
    dataList = []
    dataLineWise = list(data.split("\n"))
    del(dataLineWise[-1])
    for i in dataLineWise:
        dataList.append(tuple(map(float,i.split(", "))))
    List = []
    for i in range(len(dataList)-1):
        x1 = dataList[i][0]
        x2 = dataList[i+1][0]
        t1 = dataList[i][1]
        t2 = dataList[i+1][1]
        rate = (x2-x1)/(t2-t1)
        List.append((rate,dataList[i][1]))        
    file.close()
    FileName = directory + "AngularChangeRate\\"+str(fileNum)+ extension
    File = open(FileName,"w")
    for Data in List:
        File.write(str(Data[0])+", "+str(Data[1])+"\n")
    File.close()
```
